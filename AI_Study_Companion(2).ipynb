```python
from google import genai
```

```python
client = genai.Client(
      api_key="YOUR_API_KEY_HERE"
      )
```

```python
response = client.models.generate_content(
      model="gemini-2.5-flash",
          contents="Who was Isaac Newton?"
          )
print(response.text)
```

```python
def study_planner(subject, days):
      prompt = f"""
      Create a {days}-day study plan for {subject}.

      For each day include:

      - Topics to study
      - Approximate hours needed
      - Revision suggestions

      Keep the plan practical and beginner friendly.
      """
      response = client.models.generate_content(
                model="gemini-2.5-flash",
                contents=prompt
            )
      return response.text
```

```python
print(study_planner("Calculus", 7))
```

```python
def quiz_generator(topic, num_questions):
      prompt = f"""
      Create {num_questions} multiple-choice questions on {topic}.

      For each question provide:

      - Question
      - Four options (A, B, C, D)
      - Correct answer

      Keep the questions beginner friendly.
      """

      response = client.models.generate_content(
            model="gemini-2.5-flash",
            contents=prompt
      )
      return response.text
```

```python
print(quiz_generator("Newton's Laws", 5))
```

```python
def summarize_notes(notes):
      prompt = f"""
      Summarize the following notes.

      Provide:

      - Key concepts
      - Important points
      - Short revision notes

      Notes:

      {notes}
      """

      response = client.models.generate_content(
             model="gemini-2.5-flash",
            contents=prompt
         )
      return response.text
```

```python
notes = """
Photosynthesis is the process by which plants use sunlight to convert carbon dioxide and water into glucose and oxygen. Chlorophyll absorbs sunlight and helps carry out this process.
"""

print(summarize_notes(notes))
```

```python
def exam_preparation(subject, days_left):
      prompt = f"""
      I have {days_left} days left before my {subject} exam.

       Create:

        - Daily revision schedule
        - Important topics to focus on
        - Tips for scoring well
        - Last-minute revision advice

        Keep the plan practical and beginner friendly.
        """
      response = client.models.generate_content(
              model="gemini-2.5-flash",
               contents=prompt
           )
      return response.text
```

```python
print(exam_preparation("Physics", 5))
```

```python
def study_companion(task, topic, value):
      if task == "plan":
       return study_planner(topic, value)

      elif task == "quiz":
       return quiz_generator(topic, value)

      elif task == "exam":
        return exam_preparation(topic, value)

      else:
        return "Task not recognized."
```

```python
print(study_companion("plan", "Calculus", 7))
```

```python
print(study_companion("quiz", "Newton's Laws", 5))
```

```python
print(study_companion("exam", "Physics", 5))
```

```python
planner_agent = {
      "name": "Planner Agent",
          "purpose": "Creates personalized study plans."
          }

quiz_agent = {
              "name": "Quiz Agent",
                  "purpose": "Generates multiple-choice questions."
                  }
notes_agent = {
                      "name": "Notes Agent",
                          "purpose": "Summarizes study notes."
                          }

exam_agent = {
                              "name": "Exam Agent",
                                  "purpose": "Creates revision schedules."
                                  }
```

```python
agents = [
      planner_agent,
          quiz_agent,
              notes_agent,
                  exam_agent
                  ]

for agent in agents:
                      print(agent["name"])
                      print(agent["purpose"])
                      print()
```

```python
from google.adk.agents import Agent
```

```python
def planner_tool(subject, days):
      return study_planner(subject, days)


def quiz_tool(topic, num_questions):
      return quiz_generator(topic, num_questions)


def notes_tool(notes):
      return summarize_notes(notes)
def exam_tool(subject, days_left):
      return exam_preparation(subject, days_left)
```

```python
planner_adk_agent = Agent(
      name="PlannerAgent",
          description="Creates personalized study plans for students.",
      tools=[planner_tool]
          )

print(planner_adk_agent)
```

```python
quiz_adk_agent = Agent(
      name="QuizAgent",
          description="Generates multiple-choice questions.",
      tools=[quiz_tool]
          )
print(quiz_adk_agent)
```

```python
notes_adk_agent = Agent(
      name="NotesAgent",
          description="Summarizes study notes.",
      tools=[notes_tool]
          )
print(notes_adk_agent)
```

```python
exam_adk_agent = Agent(
      name="ExamAgent",
          description="Creates exam preparation plans.",
      tools=[exam_tool]
          )
print(exam_adk_agent)
```
